```python
import pandas as pd
```

```python
claimreal = pd.read_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/claimreal_preprocessed.csv")
claimfake = pd.read_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/claimfake_preprocessed.csv")
```

```python
claimtweets = pd.read_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/claim_tweets_preprocessed.csv")
claimreplies = pd.read_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/claim_replies_preprocessed.csv")
```

```python
claimreal.rename(columns={'Unnamed: 0	': 'index'}, inplace=True)
claimfake.rename(columns={'Unnamed: 0	': 'index'}, inplace=True)
```

```python
def addtext_claims(inputdata):
  inputdata["claimtext"] = ""
  for i in range(len(inputdata["claim_id"])):
    claim_id = inputdata["claim_id"].iloc[i]
    if inputdata["label"][i] == 1:
      inputdata["claimtext"][i] = ''.join([i for i in claimfake[claimfake.isin([claim_id]).any(axis=1)]['text'] if not i.isdigit()])
    elif inputdata["label"][i] == 0:
     inputdata["claimtext"][i] = ''.join([i for i in claimreal[claimreal.isin([claim_id]).any(axis=1)]['text'] if not i.isdigit()])
  return inputdata
```

```python
claimreplies = addtext_claims(claimreplies)
```

```python
claimreplies.to_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/Added Claim News text/replies_claimtext_added_preprocessed.csv",index=False)
```

```python
newstweets = pd.read_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/news_tweets_preprocessed.csv")
newsreplies = pd.read_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/news_replies_preprocessed.csv")
```

```python
MK_newsreal = pd.read_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/MK_newsreal_preprocessed.csv",keep_default_na = False)
MK_newsfake = pd.read_csv("/content/drive/MyDrive/Thesis/Data/Filled Datasets/Preprocessed All/MK_newsfake_preprocessed.csv",keep_default_na = False)
```

```python
newsreal.rename(columns={'Unnamed: 0	': 'index'}, inplace=True)
newsfake.rename(columns={'Unnamed: 0	': 'index'}, inplace=True)
```

```python
def addtext_news(inputdata):
  inputdata["newstext"] = ""
  for i in range(len(inputdata["news_id"])):
    news_id = inputdata["news_id"].iloc[i]
    if inputdata["label"][i] == 1:
      inputdata["newstext"][i] = ''.join([i for i in MK_newsfake[MK_newsfake.isin([news_id]).any(axis=1)]['text'] if not i.isdigit()])
    elif inputdata["label"][i] == 0:
     inputdata["newstext"][i] = ''.join([i for i in MK_newsreal[MK_newsreal.isin([news_id]).any(axis=1)]['text'] if not i.isdigit()])
  return inputdata
```

```python
MK_newstweets = addtext_news(newstweets)
MK_newsreplies = addtext_news(newsreplies)
```
